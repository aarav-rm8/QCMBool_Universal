```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('run1\program_data.csv')
data
```

```python
fig, ax1 = plt.subplots()

# Plot CPU and wall time for poly, ddsim, and aer with points
ax1.scatter(data['h'], data['cpu_time_poly'], color='blue', label='CPU Time Poly', marker='o')
ax1.scatter(data['h'], data['wall_time_poly'], color='blue', label='Wall Time Poly', marker='x')
ax1.scatter(data['h'], data['cpu_time_ddsim'], color='red', label='CPU Time DDSIM', marker='o')
ax1.scatter(data['h'], data['wall_time_ddsim'], color='red', label='Wall Time DDSIM', marker='x')
ax1.scatter(data['h'], data['cpu_time_aer'], color='green', label='CPU Time Aer', marker='o')
ax1.scatter(data['h'], data['wall_time_aer'], color='green', label='Wall Time Aer', marker='x')

# Set labels for the axes
ax1.set_xlabel('h')
ax1.set_ylabel('Time (s)')
ax1.set_title('Comparison of CPU and Wall Time')

# Create another y-axis for the value of n
ax2 = ax1.twinx()
ax2.plot(data['h'], data['n'], color='purple', label='Value of n', linestyle='dotted')
ax2.set_ylabel('Value of n')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show plot
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D


# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot CPU time for poly, ddsim, and aer with points
ax.scatter(data['h'], data['n'], data['cpu_time_poly'], color='blue', label='CPU Time Poly', marker='o')
ax.scatter(data['h'], data['n'], data['cpu_time_ddsim'], color='red', label='CPU Time DDSIM', marker='o')
ax.scatter(data['h'], data['n'], data['cpu_time_aer'], color='green', label='CPU Time Aer', marker='o')

# Set labels for the axes
ax.set_xlabel('h')
ax.set_ylabel('n')
ax.set_zlabel('CPU Time (s)')
ax.set_title('Comparison of CPU Time in 3D')

# Add legend
ax.legend()

# Show plot
plt.show()
```
